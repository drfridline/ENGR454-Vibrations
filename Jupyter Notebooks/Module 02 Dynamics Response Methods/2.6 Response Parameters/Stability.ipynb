# Stability
This notebook shows a plot of a second order system with stable, neutrally stable, and unstable responses.

The model is of the form
$$ \ddot{x} + a\dot{x} + b = 0, x(0) = x_0, \dot{x}(0) = v_0 $$

The following code block sets up the notebook and loads all necessary packages.

```python
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
%matplotlib inline
```

Use the following cell block to establish the system parameters for 3 separate systems (stable, neutrally stable, and unstable).

```python
a = [1,0,-1]
b = 10
x0 = 1
v0 = 0
```

The following code block defines a function to compute the response.
$$ x(t) = e^{\sigma t}\left[x_0\cos\omega t + \frac{v_0-x_0\sigma}{\omega}\sin\omega t\right] $$

where
$$ \sigma = -\frac{a}{2} \qquad \omega = \frac{\sqrt{4b-a^2}}{2} $$

```python
def response(a, b, x0, v0):
    sigma = -a/2
    omega = np.sqrt(4*b-a**2)
    t = np.linspace(0,10,1001)
    x = np.exp(sigma*t)*(x0*np.cos(omega*t)+(v0-x0*sigma)/omega*np.sin(omega*t))
    return t, x, sigma, omega
```

Call the function to generate the response for the three different cases.

```python
t1, x1, sigma1, omega1 = response(a[0], b, x0, v0)
t2, x2, sigma2, omega2 = response(a[1], b, x0, v0)
t3, x3, sigma3, omega3 = response(a[2], b, x0,v0)
```

Plot the responses.

```python
fig = plt.figure()
fig.suptitle('Response of Second-Order Model')
ax = fig.add_subplot(111)
ax.set_xlabel('time, $t$')
ax.set_ylabel('response, $x(t)$')
ax.grid(True)
ax.set_xlim(-0.25,4.25)
ax.set_ylim(-5,5)
plt.plot(t1,x1,label="stable")
plt.plot(t2,x2,label="neutrally stable")
plt.plot(t3,x3,label="unstable")
ax.legend()
fig.savefig('Stability.png',format='png');
```
