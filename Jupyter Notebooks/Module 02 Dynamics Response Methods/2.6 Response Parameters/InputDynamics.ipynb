```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create response plot
t = np.linspace(0,2,101)
x = 1+ np.exp(-2*t)
fig = plt.figure()
fig.suptitle('First-Order Response with Input Dynamics')
ax = fig.add_subplot(111)
ax.set_xlabel(r'time, $t$')
ax.set_ylabel(r'response, $x(t)$')
ax.grid(True)
plt.plot(t,x)
plt.plot(0,x[0],'ko')
ax.annotate(r'$x(0^+)$ = %0.2f (impulse)' % (x[0]),xy=(0,x[0]),xytext=(0.5,1.8),
            arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=160,rad=20"))

fig.savefig('InputDynamicsResponse.png',format='png');
```
