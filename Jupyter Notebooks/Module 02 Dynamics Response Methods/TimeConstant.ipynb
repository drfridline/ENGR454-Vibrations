# Time Constant for First-Order Systems
Consider the following __first-order__ system

$$ \dot{x} + ax = b. $$

## Free Response
Setting the input $b=0$, one may obtain the __free-response__ as

$$ x(t) = x(0)e^{-at}. $$

Introducing the parameter $\tau = 1/a$, provided that $a > 0$, so that

$$ \tau\dot{x} + x = b\tau \qquad\Rightarrow\qquad x(t) = x(0)e^{-t/\tau} $$

The new parameter $\tau$ has __units of time__ and is the model's __time constant__.  A plot of the response reveals some key characteristics of the time constant.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create response plot
x0 = 1
t = np.linspace(0,5,26)
x = x0*np.exp(-t)
fig = plt.figure()
fig.suptitle('Time Constant')
ax = fig.add_subplot(111)
ax.set_xlabel(r'normalized time, $t/\tau$')
ax.set_ylabel(r'normalized response, $x(t)/x(0)$')
ax.grid(True)
plt.plot(t,x);

# Annotate time constants
plt.plot(1,0.37,'ko',markersize=8)
plt.plot(4,0.02,'ko',markersize=8)

ax.annotate(r'$x(1\tau) = 0.37x(0)$',xy=(1,0.37),xytext=(2,0.6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=80,rad=20"))
ax.annotate(r'$x(4\tau) = 0.02x(0)$',xy=(4,0.02),xytext=(2.5,0.2),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=80,rad=20"))

fig.savefig('TimeConstant.png',format='png');
```

## Forced Response


```python
# Create Response Plot
xss = 1
x = xss*(1-np.exp(-t))
fig = plt.figure()
fig.suptitle('Time Constant')
ax = fig.add_subplot(111)
ax.set_xlabel(r'normalized time, $t/\tau$')
ax.set_ylabel(r'normalized response, $x(t)/x_{xx}$')
ax.grid(True)
plt.plot(t,x);

# Annotate time constants
plt.plot(1,0.63,'ko',markersize=8)
plt.plot(4,0.98,'ko',markersize=8)

ax.annotate(r'$x(1\tau) = 0.63x_{ss}$',xy=(1,0.63),xytext=(2,0.4),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=80,rad=20"))
ax.annotate(r'$x(4\tau) = 0.98x_{ss}$',xy=(4,0.98),xytext=(2,0.6),
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=80,rad=20"))

fig.savefig('TimeConstant2.png',format='png');
```
