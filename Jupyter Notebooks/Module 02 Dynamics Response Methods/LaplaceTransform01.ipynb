```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

t = np.linspace(0,5,51)
x = 10*np.exp(-3*t) + 5/3*t - 5/9 + 5/9*np.exp(-3*t)
fig = plt.figure()
fig.suptitle(r'Response for $\dot{x}+3x=5t$, for $x(0)=10$')
ax = fig.add_subplot(111)
ax.set_xlabel('time, $t$')
ax.set_ylabel('$x(t)$')
ax.grid(True)
plt.plot(t,x)
fig.savefig('LaplaceTransform01.png',format='png');
```
