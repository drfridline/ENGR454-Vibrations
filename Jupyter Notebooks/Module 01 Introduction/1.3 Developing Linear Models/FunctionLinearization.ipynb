# Linearization: Piecewise Approximation of Sine

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
# Original sine function
x = np.linspace(-1/3*np.pi, np.pi, 201)
f = np.sin(x)

# Reference points upon which to approximate function
xr = np.array([0,1/3*np.pi,2/3*np.pi])
fr = np.sin(xr)

# Linear approximations
delx = 0.75/6*np.pi
x1 = np.array([xr[0]-delx,xr[0],xr[0]+delx])
f1 = x1

x2 = np.array([xr[1]-delx,xr[1],xr[1]+delx])
f2 = 0.5*(x2-1/3*np.pi)+0.866

x3 = np.array([xr[2]-delx,xr[2],xr[2]+delx])
f3 = -0.5*(x3-2/3*np.pi)+0.866

# Plot
fig = plt.figure()
fig.suptitle(r'$f = \sin \theta$')
ax = fig.add_subplot(111)
ax.set_xlabel(r'$\theta$ [rad]')
ax.set_ylabel(r'$f(\theta)$')
ax.set_xticks([0,1/3*np.pi,2/3*np.pi])
ax.set_xticklabels([0,r'$\frac{1}{3}\pi$',r'$\frac{2}{3}\pi$'])
ax.grid(True)
ax.plot(x,f,label=r'$\sin \theta$')
ax.plot(xr,fr,'ko',markersize=8,label='reference points')
ax.plot(x1,f1,'--',label=r'$\sin(0)$')
ax.plot(x2,f2,'--',label=r'$\sin(\pi/3)$')
ax.plot(x3,f3,'--',label=r'$\sin(2\pi/3)$')
ax.legend()
fig.savefig('SineApproximation.png',format='png');
```

# Linear Approximation of Aerobee Rocket Drag Force

```python
# Define original velocity profile
rho = 0.0023
d = 1.25
CD = 0.4
A = 0.25*np.pi*d**2
v = np.linspace(0,1000,1001)
FD = 0.5*rho*A*CD*v**2

# Compute linear approximation
delv = 200
vR = 600
v1 = np.array([vR-delv,vR,vR+delv])
FD1 = 203 + 0.677*(v1-vR)

# Plot
fig = plt.figure()
fig.suptitle('Aerobee Rocket Drag Force')
ax = fig.add_subplot(111)
ax.set_xlabel(r'velocity, $v$ [ft/s]')
ax.set_ylabel(r'drag force, $F_D$ [lb$_f$]')
ax.grid(True)
ax.plot(v,FD,label=r'drag force, $F_D$')
ax.plot(v1,FD1,'--',label='Taylor series approximation');
ax.plot(v[601],FD[601],'ko',label='reference point')
ax.plot([v[0],v[1000]],[FD[0],FD[1000]],'--',label='conservative estimate')
ax.legend()
fig.savefig('AerobeeRocket.png',format='png');
```

# Graphical Representation of 3 Types of Functions
Plotting of 3 functions

1. Linear: $y = 11-6x$
2. Power: $y = 2x^{-0.5}$
3. Exponential: $y = 10\times 10^{-x}$

```python
# Define the functions
x = np.linspace(0,2,21)
y1 = 11-6*x
y2 = 2*x**(-0.5)
y3 = 10*10**(-x)

# Plot on Linear Scale
fig = plt.figure()
fig.suptitle('Rectilinear Scale')
ax = fig.add_subplot(111)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.grid(True)
ax.plot(x,y1,label='linear function')
ax.plot(x,y3,'--',label='exponential function')
ax.plot(x,y2,'--',label='power function');
ax.legend()
fig.savefig('RectilinearScale.png',format='png');
```

```python
# Plot on semilog scale
fig = plt.figure()
fig.suptitle('Rectilinear Scale')
ax = fig.add_subplot(111)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.grid(True)
ax.plot(x,y1,'--',label='linear function')
ax.plot(x,y3,label='exponential function')
ax.plot(x,y2,'--',label='power function');
ax.set_yscale('log')
ax.legend()
fig.savefig('SemilogScale.png',format='png');
```

```python
# Plot on log-log scale
fig = plt.figure()
fig.suptitle('Rectilinear Scale')
ax = fig.add_subplot(111)
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$y$')
ax.grid(True)
ax.plot(x,y1,'--',label='linear function')
ax.plot(x,y3,'--',label='exponential function')
ax.plot(x,y2,label='power function');
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()
fig.savefig('LoglogScale.png',format='png');
```
