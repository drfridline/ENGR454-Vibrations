# Developing Linear Models From Data
If we are given data on the input-output characteristics of a system element, we can first plot the data to see whether a linear model is appropriate, and if so, we can extract a suitable model.  The following problem illustrates a common engineering problem -- the estimation of the force-deflection characteristics of a cantilever support beam.

## A Cantilever Beam Deflection Model

The deflection of a cantilever beam is the distance its end moves in response to a force applied at the end.  This distance is called the **deflection** and it is the **output** variable.  The **applied force** is the **input** variable.

<img src="Cantilever.png" width=400 />

## Setup
There are a number of things we should do prior to solving this problem.  For instance we should import any  libraries and sub-libraries that we may need to use.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

Also, if you have not done so already, download the file `CantileverData.in` that was included with this Jupyter notebook. Move it into the same folder as this notebook file. Reading the data is now simple: 

```python
data = np.loadtxt("CantileverData.in")
```

Now you have all your data in one array. If you want to work with different columns, it is helpful to name them and extract them from the array. For example,

```python
f = data[:,0]    # Force data (lb)
x = data[:,1]    # Deflection data (in)
print('Force (lb):     ',f)
print('Deflection (in):',x)
```

## Plot of Data
In this section we will plot the deflection versus force data.  Note below we issue some plot formatting statements that look remarkably similar to MatLab plot formatting.  Also note that a semicolon is placed after the last formatting statement to suppress screen echoing of the last issued command.

```python
fig = plt.figure()
fig.suptitle('Beam Deflection vs Applied Force')
ax = fig.add_subplot(111)
ax.set_xlabel('Applied Force $f$ [lb]')
ax.set_ylabel('Beam Deflection $x$ [in]')
ax.grid(True)
ax.plot(f,x,'o',markersize=8,label=r'experimental data')
ax.legend();
```

## Linear Curve Fit

Based on the plot, the data seems to exhibit qualities of a **linear** model.  Common sense tells us that there must be zero beam deflection if there is no applied force, so the curve describing the data **must pass through the origin**.  We will take a **brute-force** approach to determine the fit.  Assuming a **straight-line fit** through the origin and the last data point.  The slope `m` of the curve-fit can be found as

```python
m = x[8]/f[8]
print('m =',m)
f1 = np.linspace(0,800,9)
x1 = m*f1
```

Replot with the curve fit

```python
fig = plt.figure()
fig.suptitle('Beam Deflection vs Applied Force')
ax = fig.add_subplot(111)
ax.set_xlabel('Applied Force $f$ [lb]')
ax.set_ylabel('Beam Deflection $x$ [in]')
ax.grid(True)
ax.plot(f,x,'o',markersize=8,label=r'experimental data')
ax.plot(f1,x1,'--',label=r'linear fit')
ax.legend()
fig.savefig('CantileverCurveFit.png',format='png');
```

The stiffness `k` is the inverse of the slope `m`

```python
k = 1/m
print('Beam stiffness, k =',round(k,2),'lb/in')
```
